### 1) 그래프 그리기

```python
def draw_usage_graph(df, rank = 100, target = "both", color = 'red'):
    center = [df["station_y"].mean(), df["station_x"].mean()]
    map = folium.Map(location=center, zoom_start=10)
    
    if target == "both":
        column = "total_user_count"
    elif target == "citizen":
        column = "citizen_user_count"
    elif target == "tourist":
        column = "tourist_user_count"
    else:
        print(target, " is not exist target")
        return
    df = df.sort_values(by=column, ascending = False).head(rank)
    
    for i in df.index:
        folium.CircleMarker([df.loc[i, "station_y"], df.loc[i, "station_x"]], color=color, weight = 1,
                        radius=float(df.loc[i, column])/100000*10+2, fill_color=color).add_to(map)
    return map
                                
# graph = draw_usage_graph(usage_df, rank = 300, target = "tourist", color = "red")
```
