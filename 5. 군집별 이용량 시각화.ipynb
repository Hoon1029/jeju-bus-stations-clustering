```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import folium
import datetime
import bus
import math
```

```python
def draw_usage_graph(df, rank = 100, target = "both", color = 'red'):
    center = [df["cluster_y"].mean(), df["cluster_x"].mean()]
    map = folium.Map(location=center, zoom_start=10)
    
    if target == "both":
        column = "total_user_count"
    elif target == "citizen":
        column = "citizen_user_count"
    elif target == "tourist":
        column = "tourist_user_count"
    else:
        print(target, " is not exist target")
        return
    df = df.sort_values(by=column, ascending = False).head(rank)
    
    for i in df.index:
        folium.CircleMarker([df.loc[i, "cluster_y"], df.loc[i, "cluster_x"]], color=color, weight = 1,
                        radius=float(df.loc[i, column]) ** (1.0/3.0)/100+3, fill_color=color).add_to(map)
    return map
```

```python
station_df = pd.read_csv("station_list.csv", encoding = "cp949")
cluster_df = pd.read_csv("cluster_list.csv", encoding = "cp949")
clustered_station_df = bus.create_clustered_station_df(station_df, cluster_df)
```

```python
draw_usage_graph(clustered_station_df, rank = 200, target = "tourist", color = "red")
```

```python
draw_usage_graph(clustered_station_df, rank = 200, target = "citizen", color = "blue")
```
