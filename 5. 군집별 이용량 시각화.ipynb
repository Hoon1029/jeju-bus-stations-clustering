```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import folium
import datetime
import bus
import math
```

```python
def draw_usage_graph(df, rank = 100, target = "both", color = 'red'):
    center = [df["cluster_y"].mean(), df["cluster_x"].mean()]
    map = folium.Map(location=center, zoom_start=10)
    
    if target == "both":
        column = "total_user_count"
    elif target == "citizen":
        column = "citizen_user_count"
    elif target == "tourist":
        column = "tourist_user_count"
    else:
        print(target, " is not exist target")
        return
    df = df.sort_values(by=column, ascending = False).head(rank)
    
    for i in df.index:
        folium.CircleMarker([df.loc[i, "cluster_y"], df.loc[i, "cluster_x"]], color=color, weight = 1,
                        radius=float(df.loc[i, column]) ** (1.0/3.0)/100+3, fill_color=color).add_to(map)
    return map
```

```python
station_df = bus.create_station_df()
cluster_df = bus.create_cluster_df()
clustered_station_df = bus.create_clustered_station_df(station_df, cluster_df)
```

### 관광객이 1000번 이상 이용한 정류장

```python
df = clustered_station_df[clustered_station_df["tourist_user_count"]>1000]
draw_usage_graph(df, rank = 10000, target = "tourist", color = "red")
```

### 도민이 3000번 이상 이용한 정류장

```python
# 1000번 이상 방문한 
df = clustered_station_df[clustered_station_df["citizen_user_count"]>3000]
draw_usage_graph(df, rank = 10000, target = "citizen", color = "blue")
```
