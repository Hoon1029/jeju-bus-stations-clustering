```python
def extract_by_period_df(user_df, min_day, max_day):
    user_df = user_df.loc[:, ["tour_period", "user_id"]]
    user_df["tour_period"] = user_df["tour_period"].apply(lambda x:str(x))
    user_df["tour_period"] = user_df["tour_period"].apply(lambda x:x.split(" ")[0])
    user_df["tour_period"] = user_df["tour_period"].apply(lambda x:int(x))
    user_df = user_df[min_day <= user_df["tour_period"]]
    user_df = user_df[user_df["tour_period"] <= max_day]
    return user_df
```
