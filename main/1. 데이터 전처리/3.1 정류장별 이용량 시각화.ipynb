```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import folium
import datetime

import bus.analyzer as anz
```

### 1) 그래프 그리기

```python
def draw_usage_graph(df, rank = 100, target = "both", color = 'red'):
    center = [df["station_longitude"].mean(), df["station_latitude"].mean()]
    map = folium.Map(location=center, zoom_start=10)
    
    if target == "both":
        column = "total_usaget"
    elif target == "citizen":
        column = "regident_geton_usage"
    elif target == "tourist":
        column = "tour_geton_usage"
    else:
        print(target, " is not exist target")
        return
    df = df.sort_values(by=column, ascending = False).head(rank)
    
    for i in df.index:
        folium.CircleMarker([df.loc[i, "station_latitude"], df.loc[i, "station_longitude"]], color=color, weight = 1,
                        radius=float(df.loc[i, column]) ** (1.0/3.0)/100+3, fill_color=color).add_to(map)
    return map
                                
```
